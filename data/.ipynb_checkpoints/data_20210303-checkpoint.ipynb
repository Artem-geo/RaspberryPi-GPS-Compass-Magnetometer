```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# rotation
df_mg_rot = pd.read_csv('./20210303/20210303_104454_MG_c.txt', sep=' ', header=0)
df_az_rot = pd.read_csv('./20210303/20210303_104447_AZ.txt', sep=' ', header=0)
```

```python
df_mg_rot.sort_values(by='GPST', axis=0, inplace=True)
df_az_rot.sort_values(by='GPST', axis=0, inplace=True)

df_rot = pd.merge_asof(df_mg_rot, df_az_rot, on='GPST', direction='nearest')
df_rot.sort_values(by='GPST', axis=0, inplace=True)
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0,0,1,1], projection='polar')
ax.set_ylim(52700, 52780)
ax.scatter(df_rot['AZ'], df_rot['FIELD'])
```

```python
df_mg_surv = pd.read_csv('./20210303/20210303_114424_MG_c.txt', sep=' ', header=0)
df_az_surv = pd.read_csv('./20210303/20210303_114417_AZ.txt', sep=' ', header=0)
df_mg_surv['FIELD'] = df_mg_surv['FIELD'].apply(lambda row: int(row*1000))
```

```python
df_mg_from_pos = pd.read_csv('./20210303/2021-03-03_23-20-45.txt', sep=' ', header=0)
df_mg_from_pos = df_mg_from_pos[['FIELD', 'DATE', 'TIME']]
df_mg_from_pos['FIELD'] = df_mg_from_pos['FIELD'].apply(lambda row: int(row))
df_mg_from_pos['GPST_s'] = np.nan
df_mg_from_pos['DATE(UTC)_s'] = np.nan
df_mg_from_pos['TIME(UTC)_s'] = np.nan
df_mg_from_pos['FIELD_s'] = np.nan
```

```python
for i in range(df_mg_from_pos.shape[0]):
        if (df_mg_from_pos['FIELD'][i] == df_mg_surv['FIELD'][0]):
            df_mg_from_pos['GPST_s'][i] = df_mg_surv['GPST'][0]
            df_mg_from_pos['DATE(UTC)_s'][i] = df_mg_surv['DATE(UTC)'][0]
            df_mg_from_pos['TIME(UTC)_s'][i] = df_mg_surv['TIME(UTC)'][0]
            df_mg_from_pos['FIELD_s'][i] = df_mg_surv['FIELD'][0]
            
            df_mg_surv.drop(axis=0, index=0, inplace=True)
            df_mg_surv.reset_index(inplace=True)
            df_mg_surv.drop('index', axis=1, inplace=True)

        if df_mg_surv.shape[0] == 0:
            break
```

```python
df_mg_from_pos['GPST_s'] = df_mg_from_pos['GPST_s'].interpolate(method='linear')
```

```python
df_mg_from_pos.drop(['DATE', 'TIME'], axis=1, inplace=True)
```

```python
df_mg_from_pos.drop(['DATE(UTC)_s', 'TIME(UTC)_s', 'FIELD_s'], axis=1, inplace=True)
```

```python
df_mg = df_mg_from_pos.dropna()
```

```python
df_mg['FIELD'] = df_mg['FIELD']/1000
```

```python
df_mg['GPST'] = df_mg['GPST_s']
```

```python
df_mg.drop('GPST_s', axis=1, inplace=True)
```

```python
df_mg.sort_values(by='GPST', axis=0, inplace=True)
df_az_surv.sort_values(by='GPST', axis=0, inplace=True)
```

```python
df_res = pd.merge_asof(df_mg, df_az_surv, on='GPST', direction='nearest')
```

```python
plt.figure(figsize=(12,5))
plt.plot(df_res['GPST'], df_res['AZ'])
```

```python
df_res.to_csv('./20210303/20210303_flight.txt', sep=' ')
```
