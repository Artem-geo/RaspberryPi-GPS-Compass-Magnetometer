### Вращение

```python
def sort_and_merge(mg, az):
    mg.sort_values(by='GPST', axis=0, inplace=True)
    az.sort_values(by='GPST', axis=0, inplace=True)

    mgaz = pd.merge_asof(mg, az, on='GPST', direction='nearest')
    mgaz.sort_values(by='AZ', axis=0, inplace=True)

    mgaz['AZ_rad'] = (mgaz['AZ'] * np.pi)/180
```

```python
# first quarter
mg1 = pd.read_csv('./20200305/20210305_073020_MG_c.txt', sep=' ', header=0)
az1 = pd.read_csv('./20200305/20210305_073016_AZ.txt', sep=' ', header=0)
mgaz1 = sort_and_merge(mg1, az1)

# second quarter
mg2 = pd.read_csv('./20200305/20210305_073213_MG_c.txt', sep=' ', header=0)
az2 = pd.read_csv('./20200305/20210305_073209_AZ.txt', sep=' ', header=0)
mgaz2 = sort_and_merge(mg2, az2)

# third quarter
mg3 = pd.read_csv('./20200305/20210305_073357_MG_c.txt', sep=' ', header=0)
az3 = pd.read_csv('./20200305/20210305_073354_AZ.txt', sep=' ', header=0)
mgaz3 = sort_and_merge(mg3, az3)

#fourth quarter
mg4 = pd.read_csv('./20200305/20210305_073526_MG_c.txt', sep=' ', header=0)
az4 = pd.read_csv('./20200305/20210305_073522_AZ.txt', sep=' ', header=0)
mgaz4 = sort_and_merge(mg4, az4)

mgaz = [mgaz1, mgaz2, mgaz3, mgaz4]


# fig = plt.figure(figsize=(5,5))
# ax = fig.add_axes([0,0,1,1], projection='polar')
# ax.set_ylim(50000, 54000)
# colors = ['red', 'green', 'blue', 'yellow']
# for i, c in enumerate(colors):
#     ax.scatter(mgaz[i]['AZ_rad'], mgaz[i]['FIELD'], c=colors[i])
```
