```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cufflinks as cf
```

```python
cf.go_offline()
```

```python
df_mg = pd.read_csv('20210301_134753_MG_c.txt', sep=' ', header=0)
df_az = pd.read_csv('20210301_140411_AZ.txt', sep=' ', header=0)
```

```python
df_mg.sort_values(by='GPST', axis=0, inplace=True)
df_az.sort_values(by='GPST', axis=0, inplace=True)
```

```python
df = pd.merge_asof(df_mg, df_az, on='GPST', direction='nearest')
```

```python
df['AZ']
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_axes([0,0,1,1], projection='polar')
ax.set_ylim(45000, 58000)
ax.scatter(df['AZ'], df['FIELD'])
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0,0,1,1])
ax.scatter(df_az['GPST'], df_az['AZ'])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = np.array(df['AZ']).reshape(-1, 1)
y = np.array(df['FIELD']).reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
pred = lr.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.mean_squared_error(pred, y_test)
```
